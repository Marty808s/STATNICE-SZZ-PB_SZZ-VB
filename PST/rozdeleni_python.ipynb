```python
import numpy as np
import matplotlib.pyplot as plt

def getNormalniRozdeleni(x, mean, rozp):
    return (1 / (rozp * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / rozp) ** 2)

x = np.linspace(0, 10, 200) 
y = getNormalniRozdeleni(x, 5, 1)

plt.plot(x, y, "r-")
plt.title("Normální rozdělení N(0,1)")
plt.show()
```
